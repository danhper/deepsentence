```python
correspondence = articles.copy()
n_col_articles = articles.shape[1]
for i, article_id in enumerate(sources['article_id']):
    row = np.where(correspondence['id']==article_id)[0]
    if len(row):
        column = 'article%d' % (correspondence.iloc[row[0]].notnull().sum() - n_col_articles + 1)
        correspondence.ix[row[0], column] = sources.ix[i, 'content']
correspondence = correspondence.drop('id', axis=1)
```
